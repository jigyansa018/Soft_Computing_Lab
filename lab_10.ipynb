## Q2 - Implement the logic gate OR using neural network learning algorithm (take random values).

```python
import numpy as np

# Step function (activation)
def step(x):
    return 1 if x >= 0 else 0

# Perceptron training for OR gate
def train_or_gate(X, Y, lr=0.1, epochs=10):
    w = np.random.rand(2)  # Initialize weights
    b = np.random.rand(1)  # Initialize bias

    for e in range(epochs):
        total_error = 0
        for i in range(len(X)):
            net = np.dot(X[i], w) + b
            y_pred = step(net)
            error = Y[i] - y_pred

            # Update weights and bias
            w += lr * error * X[i]
            b += lr * error

            total_error += abs(error)

        print(f"Epoch {e+1}: Weights={np.round(w,1)}, Bias={np.round(b,1)}, Error={total_error}")
        if total_error == 0:
            break

    return w, b

# Input data (truth table)
X = np.array([[0,0],
              [0,1],
              [1,0],
              [1,1]])
Y_or = np.array([0,1,1,1])

print("Training OR gate:")
w_or, b_or = train_or_gate(X, Y_or)

# Testing
def test_or_gate(X, w, b):
    print("\nOR Gate Results:")
    for x in X:
        y = step(np.dot(x, w) + b)
        print(f"{x} => {y}")

test_or_gate(X, w_or, b_or)
```
