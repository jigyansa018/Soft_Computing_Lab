## Program on AND function using McCulloch-Pitts with weights w1 = 1, w2 = 1 and threshold = 2.

```python
# McCulloch-Pitts model for AND gate
import pandas as pd

# Inputs and weights
inputs = [(0, 0), (0, 1), (1, 0), (1, 1)]
w1, w2 = 1, 1
threshold = 2

def mcculloch_pitts_AND(x1, x2):
    net_input = w1 * x1 + w2 * x2
    return 1 if net_input >= threshold else 0

# Compute output
results = []
for x1, x2 in inputs:
    y = mcculloch_pitts_AND(x1, x2)
    results.append([x1, x2, y])

# Display as table
df = pd.DataFrame(results, columns=["Input 1", "Input 2", "Output (AND)"])
df
```

## Program using McCulloch-Pitts Neuron Model for AND, OR, and NAND gates.

```python
import pandas as pd

# All possible binary inputs
inputs = [(0, 0), (0, 1), (1, 0), (1, 1)]

def mcculloch_pitts(x1, x2, w1, w2, threshold):
    net_input = w1*x1 + w2*x2
    return 1 if net_input >= threshold else 0

# Define functions for each gate
def AND_gate(x1, x2):
    return mcculloch_pitts(x1, x2, 1, 1, 2)

def OR_gate(x1, x2):
    return mcculloch_pitts(x1, x2, 1, 1, 1)

def NAND_gate(x1, x2):
    return 1 - mcculloch_pitts(x1, x2, 1, 1, 2)

# Generate truth tables
results = []
for x1, x2 in inputs:
    results.append([x1, x2, AND_gate(x1, x2), OR_gate(x1, x2), NAND_gate(x1, x2)])

# Display as DataFrame
df = pd.DataFrame(results, columns=["Input 1", "Input 2", "AND", "OR", "NAND"])
df
```
