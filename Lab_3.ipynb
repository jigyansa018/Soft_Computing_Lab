### Differnet oprations on fuzzy sets:

```python
import numpy as np
import matplotlib.pyplot as plt

# Universe of discourse
X = np.arange(1, 9)

# Fuzzy sets A and B
A = np.array([0, 0, 0.8, 0, 1, 0.6, 0, 0])
B = np.array([0, 0, 0.7, 1, 0, 0.5, 0, 0])

# Complement of A and B
compA = 1 - A
compB = 1 - B

# --- Plotting ---
plt.figure(figsize=(10, 5))

# Plot A and B
plt.subplot(1, 2, 1)
plt.plot(X, A, 'r-o', label="A")
plt.plot(X, B, 'b-s', label="B")
plt.ylim([0, 1.1])
plt.title("Fuzzy Sets A and B")
plt.xlabel("X")
plt.ylabel("Membership μ(x)")
plt.legend()

# Plot complements
plt.subplot(1, 2, 2)
plt.plot(X, compA, 'm-o', label="~A")
plt.plot(X, compB, 'c-s', label="~B")
plt.ylim([0, 1.1])
plt.title("Complements of A and B")
plt.xlabel("X")
plt.ylabel("Membership μ(x)")
plt.legend()

plt.tight_layout()
plt.show()
```

### T-norm  oprations

```python
import numpy as np
import matplotlib.pyplot as plt

# Universe of discourse
X = np.arange(1, 9)

# Fuzzy sets A and B
A = np.array([0, 0, 0.8, 0, 1, 0.6, 0, 0])
B = np.array([0, 0, 0.7, 1, 0, 0.5, 0, 0])

# --- T-norm Operations ---
Tmin = np.minimum(A, B)                      # Minimum T-norm
Tap = A * B                                  # Algebraic Product
Tbp = np.maximum(0, A + B - 1)               # Bounded Product
Tdp = np.array([a if b == 1 else (b if a == 1 else 0) 
                for a, b in zip(A, B)])      # Drastic Product

# --- Plotting ---
plt.figure(figsize=(12, 6))

# Original sets
plt.subplot(1, 2, 1)
plt.plot(X, A, 'r-o', label="A")
plt.plot(X, B, 'b-s', label="B")
plt.ylim([0, 1.1])
plt.title("Fuzzy Sets A and B")
plt.xlabel("X")
plt.ylabel("Membership μ(x)")
plt.legend()

# T-norm operations
plt.subplot(1, 2, 2)
plt.plot(X, Tmin, 'g-o', label="Tmin (min)")
plt.plot(X, Tap, 'm-s', label="Algebraic Product")
plt.plot(X, Tbp, 'c-^', label="Bounded Product")
plt.plot(X, Tdp, 'y-d', label="Drastic Product")
plt.ylim([0, 1.1])
plt.title("T-norm Operations")
plt.xlabel("X")
plt.ylabel("Membership μ(x)")
plt.legend()

plt.tight_layout()
plt.show()
```

### T Co-norm oprations/S-norm oprations

```python
import numpy as np
import matplotlib.pyplot as plt

# Universe of discourse
X = np.arange(1, 9)

# Fuzzy sets A and B
A = np.array([0, 0, 0.8, 0, 1, 0.6, 0, 0])
B = np.array([0, 0, 0.7, 1, 0, 0.5, 0, 0])

# --- S-norm (T-conorm) Operations ---
Smax = np.maximum(A, B)                       # Maximum (union)
Sap  = A + B - (A * B)                        # Algebraic Sum
Sbs  = np.minimum(1, A + B)                   # Bounded Sum
Sds  = np.array([a if b == 0 else (b if a == 0 else 1) 
                 for a, b in zip(A, B)])      # Drastic Sum 

# --- Plotting ---
plt.figure(figsize=(12, 6))

# Original sets
plt.subplot(1, 2, 1)
plt.plot(X, A, 'r-o', label="A")
plt.plot(X, B, 'b-s', label="B")
plt.ylim([0, 1.1])
plt.title("Fuzzy Sets A and B")
plt.xlabel("X")
plt.ylabel("Membership μ(x)")
plt.legend()

# S-norm operations
plt.subplot(1, 2, 2)
plt.plot(X, Smax, 'g-o', label="Maximum (Smax)")
plt.plot(X, Sap, 'm-s', label="Algebraic Sum")
plt.plot(X, Sbs, 'c-^', label="Bounded Sum")
plt.plot(X, Sds, 'y-d', label="Drastic Sum")
plt.ylim([0, 1.1])
plt.title("S-norm (T-conorm) Operations")
plt.xlabel("X")
plt.ylabel("Membership μ(x)")
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define x domain
x = np.linspace(0, 10, 500)

# Membership functions
mu_A = x / (x + 1)
mu_B = 2 ** (-x)

# Complements
mu_Ac = 1 - mu_A
mu_Bc = 1 - mu_B

# Plot A and B
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(x, mu_A, label="A: x/(x+1)", color="blue")
plt.plot(x, mu_B, label="B: 2^(-x)", color="red")
plt.title("Fuzzy Sets A and B")
plt.xlabel("x")
plt.ylabel("Membership Value")
plt.legend()
plt.grid(True)

# Plot complements
plt.subplot(1, 2, 2)
plt.plot(x, mu_Ac, label="A complement", linestyle="--", color="blue")
plt.plot(x, mu_Bc, label="B complement", linestyle="--", color="red")
plt.title("Complements of A and B")
plt.xlabel("x")
plt.ylabel("Membership Value")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
