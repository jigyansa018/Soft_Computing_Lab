### Universe of Discourse

```python
# Universe of discourse
X = [1, 2, 3, 4, 5, 6, 7, 8]

# Define fuzzy sets as dictionaries {element: membership}
A = {x: 0 for x in X}
B = {x: 0 for x in X}

# Populate fuzzy sets
A[3], A[5], A[6] = 0.8, 1.0, 0.6
B[3], B[4], B[6] = 0.7, 1.0, 0.5

# Complement of A
A_comp = {x: 1 - A[x] for x in X}

# Union (max)
A_union_B = {x: max(A[x], B[x]) for x in X}

# Intersection (min)
A_inter_B = {x: min(A[x], B[x]) for x in X}

# Algebraic product
A_B = {x: A[x] * B[x] for x in X}

# Power A^3
A_power3 = {x: A[x] ** 3 for x in X}

# Scalar multiplication 0.6 * B
scalar_B = {x: min(1, 0.6 * B[x]) for x in X}

# Concentration (A^2)
Con_A = {x: A[x] ** 2 for x in X}

# Dilation (A^0.5)
Dil_A = {x: A[x] ** 0.5 for x in X}

# Display results
def print_fuzzy_set(name, fuzzy_set):
    print(f"{name}:")
    for x in X:
        if fuzzy_set[x] > 0:
            print(f"  {fuzzy_set[x]:.2f} / {x}")
    print()

print_fuzzy_set("A", A)
print_fuzzy_set("B", B)
print_fuzzy_set("~A", A_comp)
print_fuzzy_set("A U B", A_union_B)
print_fuzzy_set("A ∩ B", A_inter_B)
print_fuzzy_set("A * B (Algebraic Product)", A_B)
print_fuzzy_set("A^3", A_power3)
print_fuzzy_set("0.6 * B", scalar_B)
print_fuzzy_set("CON(A) = A^2", Con_A)
print_fuzzy_set("DIL(A) = A^0.5", Dil_A)
```

### The concept of A subset B

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Universe of discourse
X = np.arange(1, 9)

# Fuzzy subsets A and B
A = np.array([0, 0, 0.8, 0, 1, 0.6, 0, 0])
B = np.array([0, 0, 0.7, 1, 0, 0.5, 0, 0])

compA = 1 - A
unionAB = np.maximum(A, B)
interAB = np.minimum(A, B)

# --- Plotting like classroom slides ---
plt.figure(figsize=(10, 8))

# (a) Fuzzy sets A and B
plt.subplot(2, 2, 1)
plt.plot(X, A, 'r-o', label="A")
plt.plot(X, B, 'b-s', label="B")
plt.title("Fuzzy Sets A and B")
plt.ylim([0, 1.1])
plt.legend()

# (b) Complement of A
plt.subplot(2, 2, 2)
plt.plot(X, compA, 'k-o')
plt.title("Fuzzy Set NOT A")
plt.ylim([0, 1.1])

# (c) Union (A ∪ B)
plt.subplot(2, 2, 3)
plt.plot(X, unionAB, 'g-o')
plt.title("Fuzzy Set A OR B")
plt.ylim([0, 1.1])

# (d) Intersection (A ∩ B)
plt.subplot(2, 2, 4)
plt.plot(X, interAB, 'm-o')
plt.title("Fuzzy Set A AND B")
plt.ylim([0, 1.1])

plt.tight_layout()
plt.show()
```

```python
# Universe of discourse
X = np.arange(1, 9)

# Fuzzy subsets A and B (arrays indexed by X)
A = np.array([0, 0, 0.8, 0, 1, 0.6, 0, 0])
B = np.array([0, 0, 0.7, 1, 0, 0.5, 0, 0])

# --- Operations ---
compA = 1 - A              # Complement
unionAB = np.maximum(A, B) # Union
interAB = np.minimum(A, B) # Intersection
A3 = A**3                  # Power (A^3)
B06 = 0.6 * B              # Scalar multiplication
conA = A**2                # Concentration
dilA = np.sqrt(A)          # Dilation

# --- Plotting ---
fig, axs = plt.subplots(3, 3, figsize=(12, 10))

# Original sets
axs[0,0].stem(X, A, linefmt='r-', markerfmt='ro', basefmt=" ")
axs[0,0].stem(X, B, linefmt='b-', markerfmt='bo', basefmt=" ")
axs[0,0].set_ylim(0, 1.1)
axs[0,0].set_title("Original Sets A (red) and B (blue)")

# Complement of A
axs[0,1].stem(X, compA, linefmt='k-', markerfmt='ko', basefmt=" ")
axs[0,1].set_ylim(0, 1.1)
axs[0,1].set_title("Complement of A (~A)")

# Union
axs[0,2].stem(X, unionAB, linefmt='g-', markerfmt='go', basefmt=" ")
axs[0,2].set_ylim(0, 1.1)
axs[0,2].set_title("Union (A ∪ B)")

# Intersection
axs[1,0].stem(X, interAB, linefmt='m-', markerfmt='mo', basefmt=" ")
axs[1,0].set_ylim(0, 1.1)
axs[1,0].set_title("Intersection (A ∩ B)")

# A^3
axs[1,1].stem(X, A3, linefmt='c-', markerfmt='co', basefmt=" ")
axs[1,1].set_ylim(0, 1.1)
axs[1,1].set_title("A^3")

# 0.6B
axs[1,2].stem(X, B06, linefmt='b-', markerfmt='bo', basefmt=" ")
axs[1,2].set_ylim(0, 1.1)
axs[1,2].set_title("0.6B")

# Concentration
axs[2,0].stem(X, conA, linefmt='r-', markerfmt='ro', basefmt=" ")
axs[2,0].set_ylim(0, 1.1)
axs[2,0].set_title("Concentration (A^2)")

# Dilation
axs[2,1].stem(X, dilA, linefmt='g-', markerfmt='go', basefmt=" ")
axs[2,1].set_ylim(0, 1.1)
axs[2,1].set_title("Dilation (A^0.5)")

# Last panel with text
axs[2,2].axis('off')
axs[2,2].text(0.1, 0.5, "Graphs of Fuzzy Set Operations", fontsize=12, fontweight="bold")

plt.tight_layout()
plt.show()
```
