### Membership Function

```python
#Triangular Membership Function
import numpy as np
import matplotlib.pyplot as plt
```

```python
def triangular_mf(x, a, b, c):
    return np.maximum(np.minimum((x - a) / (b - a), (c - x) / (c - b)), 0)
```

```python
x = np.linspace(0, 100, 500)
a, b, c = 20, 60, 80
y = triangular_mf(x, a, b, c)
```

```python
plt.plot(x, y, 'k')
plt.title("Triangular MF")
plt.xlabel("x")
plt.ylabel("Membership Grades")
plt.xlim(0, 100)
plt.ylim(0, 1.1)
plt.grid(True)
plt.show()
```

```python
#Trapezoidal Membership Function
import numpy as np
import matplotlib.pyplot as plt
```

```python
def trapezoidal_mf(x, a, b, c, d):
    return np.maximum(
        np.minimum(
            np.minimum((x - a) / (b - a), 1),
            (d - x) / (d - c)
        ),
        0
    )
```

```python
x = np.linspace(0, 100, 500)
```

```python
a, b, c, d = 10, 20, 60, 95
y = trapezoidal_mf(x, a, b, c, d)
```

```python
plt.plot(x, y, 'k')
plt.title("Trapezoidal MF")
plt.xlabel("x")
plt.ylabel("Membership Grades")
plt.xlim(0, 100)
plt.ylim(0, 1.1)
plt.grid(True)
plt.show()
```

```python
#Gussian MF
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gaussian_mf(x, c, sigma):
    return np.exp(-0.5 * ((x - c) / sigma) ** 2)
```

```python
x = np.linspace(0, 100, 500)
```

```python
c, sigma = 50, 20
y = gaussian_mf(x, c, sigma)
```

```python
plt.plot(x, y, 'k')
plt.title("Gaussian MF")
plt.xlabel("x")
plt.ylabel("Membership Grades")
plt.xlim(0, 100)
plt.ylim(0, 1.1)
plt.grid(True)
plt.show()
```

```python
#Generalized Bell Membership Function
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gbell_mf(x, a, b, c):
    return 1 / (1 + np.abs((x - c) / a) ** (2 * b))
```

```python
x = np.linspace(0, 100, 500)
```

```python
a, b, c = 20, 4, 50
y = gbell_mf(x, a, b, c)
```

```python
plt.plot(x, y, 'k')
plt.title("Generalized Bell MF")
plt.xlabel("x")
plt.ylabel("Membership Grades")
plt.xlim(0, 100)
plt.ylim(0, 1.1)
plt.grid(True)
plt.show()
```
